```python
import numpy as np
import pandas as pd

df = pd.read_csv(
    "e_commerce.csv",
    dtype={
        "order_id": np.int16,
        "quantity": np.int8,
        "shipping_days": np.int8,
    },
    true_values=['Yes'],
    false_values=['No'],
    index_col=0,
)
df['order_date'] = pd.to_datetime(df['order_date'])
df.head()
```

```python
df.loc[df["return_status"]==True,["product_name"]]
```

```python
unit_price_after_discount = np.round(df['unit_price'] * (1 - df['discount_pct']), 2)

df.insert(7, "unit_price_after_discount", unit_price_after_discount)
df.insert(8, "total_price_after_discount", unit_price_after_discount * df['quantity'])
```

```python
df.insert(6, "total_price", df['unit_price'] * df['quantity'])
```

```python
df.head()
```

```python
pd.DataFrame({"Mean price before discount": [df["total_price"].mean()], "Mean price after discount": [df["total_price_after_discount"].mean()]}, index=["Mean price"])
```

```python
LE=df["total_price_after_discount"]*47.77
df.insert(9, "total_price_after_discount_in_LE", LE)
```
